```python
import transformers
# Importa a biblioteca transformers, que é usada para trabalhar com modelos de linguagem


# pipeline serve para agrupar todas as funções necessárias da tarefa
# para processar a entrada e gerar a saída
gerbo = transformers.pipeline(
    task="text-generation",  # Define a tarefa como geração de texto
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",  # Modelo TinyLlama para geração de texto
    device="cuda"  # Especifica o uso de GPU (cuda); se não disponível, use 'cpu'
)
```

```python
# Exibe informações sobre o pipeline de geração de texto
gerbo
```

```python
# modelo escolhinho: 22 blocos de processamento
# faz normalização
# suporta até 2048 tokens de input
# embedding de 32000 tokens: ele conhece 32k palavras
# tarefa dele: recebe um input (texto que será codificado em tokens com o tokenizer),
# e retorna a probabilidade de cada uma das 32k palavras de ser a próxima

gerbo.model
# Exibe detalhes sobre o modelo de linguagem carregado no pipeline
```

```python
def analisa_e_processa_prompt(modelo, prompt):
  print("="*80)  # Exibe uma linha de separação para organizar a saída

  # Codifica o prompt em tokens
  tokens = modelo.tokenizer.encode(prompt, return_tensors="pt")

  # Exibe o prompt original e seus tokens codificados
  print(f"- prompt original: {prompt}")
  print(f"- token ids:{tokens}")
  print(f"- # de tokens:{len(tokens)}")

  # Decodifica os tokens de volta para texto
  decoded = modelo.tokenizer.convert_ids_to_tokens(tokens[0])
  print(f"- tokens decodificados: {decoded}")

  # Exibe os primeiros tokens do vocabulário do modelo
  first_tokens = modelo.tokenizer.convert_ids_to_tokens(range(5))
  print(f"- primeiros tokens: {first_tokens}")

  # Gera o texto a partir do prompt usando temperatura para controlar a criatividade
  # 'do_sample=True' permite que o modelo gere variações mais criativas
  output = modelo(prompt, do_sample=True, temperature=0.7)
  print(f"- output do modelo: {output}")
  print(f"- texto gerado efetivamente: {output[0]['generated_text']}")

  # Exibe outra linha de separação
  print("="*80)
  return

# Chama a função para analisar o prompt "Once upon a time" e gerar texto com base nele
analisa_e_processa_prompt(gerbo, "Once upon a time")
```
